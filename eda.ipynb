```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import matplotlib as mpl

# 1) Disable LaTeX parsing to avoid '$' or special-character issues
mpl.rcParams['text.usetex'] = False
mpl.rcParams['axes.unicode_minus'] = False

%matplotlib inline
sns.set_style("whitegrid")

# Load your data
df = pd.read_csv('clean_tabular_data.csv')

# Replace placeholder '--' with NaN if necessary
df.replace('--', np.nan, inplace=True)

print("Data shape:", df.shape)
df.info()
display(df.head())

# Missing values
print("\nMissing values per column:")
print(df.isnull().sum())

msno.bar(df)
plt.title("Missing Values - Bar Plot")
plt.show()

# EXAMPLE: If msno.heatmap fails, wrap in try-except
try:
    msno.heatmap(df)
    plt.title("Missing Values - Heatmap")
    plt.show()
except ValueError as e:
    print(f"Heatmap error: {e}")

# Basic stats
display(df.describe(include='all'))

# Separate numeric vs. categorical
numeric_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(include=['object','category']).columns

# Numeric distribution
for col in numeric_cols:
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    sns.histplot(df[col], kde=True, color='blue')
    plt.title(f"{col} Distribution")

    plt.subplot(1,2,2)
    sns.boxplot(x=df[col], color='orange')
    plt.title(f"{col} Boxplot")

    plt.tight_layout()
    plt.show()

# Categorical distribution (skip huge columns)
for col in categorical_cols:
    unique_vals = df[col].nunique()
    if unique_vals > 20:
        print(f"Skipping {col} (too many unique values: {unique_vals})")
        continue

    plt.figure(figsize=(8,4))
    sns.countplot(x=df[col], palette='viridis')
    plt.title(f"Count Plot - {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Check skew
skews = df[numeric_cols].skew().sort_values(ascending=False)
print("\nSkewness:\n", skews)

# Correlation
corr = df[numeric_cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Pairplot (can be large)
sns.pairplot(df[numeric_cols], diag_kind='kde', corner=True)
plt.suptitle("Pairwise Plots", y=1.02)
plt.show()
```
